```python
#here we are finding the null values
# Create a list of dataset names
dataset_names = ["data1", "data2", "data3", "data4", "data5"]

# Iterate through the dataset names and find null values
for dataset_name in dataset_names:

    data = globals()[dataset_name]
    null_count = data.isnull().sum()
    print(f"Null values in {dataset_name}:\n{null_count}")
```

```python
# Create a dictionary of datasets
datasets = {
    "data1": data1,
    "data2": data2,
    "data3": data3,
    "data4": data4,
    "data5": data5
}

# Iterate through the dictionary and find null values
for dataset_name, dataset in datasets.items():
    null_count = dataset.isnull().sum()
    print(f"Null values in {dataset_name}: {null_count}")
```

```python
dataset_names = ["data1", "data2", "data3", "data4", "data5"]

for dataset_name in dataset_names:
  data = globals()[dataset_name]
  info = data.info()
  print(f"info of {dataset_name}: {info}")
```

```python
dataset_names = ["data1", "data2", "data3", "data4", "data5"]

for dataset_name in dataset_names:
    data = globals()[dataset_name]

    # Check if the "Price" column exists in the dataset
    if "Price" in data.columns:
        # Convert the "Price" column to int
        data["Price"] = data["Price"].str.replace(",", "").astype(int)
        print(f"Converted 'Price' column to int in {dataset_name}")

    # Check if the "DepartingTime" column exists in the dataset
    if "DepartingTime" in data.columns:
        # Create a "DepartingHour" column and set its values based on splitting "DepartingTime"
        data["DepartingHour"] = data["DepartingTime"].str.split(":").str[0].astype(int)
        print(f"Created 'DepartingHour' column in {dataset_name}")
    else:
        print(f"'DepartingTime' column not found in {dataset_name}")
```

```python
#coverting the departingtime into the int
dataset_names = ["data1", "data2", "data3", "data4", "data5"]

for dataset_name in dataset_names:
    data = globals()[dataset_name]

    # Check if the "DepartingTime" column exists in the dataset
    if "DepartingTime" in data.columns:
        data["DepartingTime"] = data["DepartingTime"].astype(int, errors="ignore")
        print(f"Converted 'DepartingTime' column to int in {dataset_name}")
    else:
        print(f"'DepartingTime' column not found in {dataset_name}")
```
